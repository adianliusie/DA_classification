### This Notebook aims to run experiments to see the focus of the baseline seq2seq models for DA classification.

It was hypothesised that all the encoder has all utterance seperation information, it's hard for the decoder to know what the current utterance has to be (due to complex information overload) and this notebook looks at how much focus the decoder has for each decision.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

from types import SimpleNamespace
import scipy

from src.eval_handler import EvalHandler
from src.config import config
```

```python
def saliency_entropy(dist:list):
    dist = [abs(i) for i in dist]
    dist = np.array(dist)/sum(dist)
    return scipy.stats.entropy(dist, base=2)
```

```python
def plot_focus(conv_list):
    sns.set_theme()
    plt.rcParams["figure.figsize"] = 10, 5

    for focus, entropy, certainty in conv_list:    
        x = range(1, len(focus)+1)
        plt.plot(x, focus)

    plt.xlabel('utterance number')
    plt.ylabel('focus of current utterance')
    plt.show()
    
def plot_entropy(conv_list):
    sns.set_theme()
    plt.rcParams["figure.figsize"] = 10, 5

    for focus, entropy, certainty in conv_list:    
        x = range(1, len(entropy)+1)
        plt.plot(x, entropy)

    plt.xlabel('utterance number')
    plt.ylabel('entropy of focus')
    plt.show()
```

```python
E = EvalHandler('seq2seq/bs4_lr-5_e80')

eval_path = f"{config.base_dir}/data/swda/standard/dev.json"
l_path = f"{config.base_dir}/data/swda/standard/labels.json"

args = {'test_path':eval_path,
        'bsz':1, 
        'override':False,
        'label_path':l_path, 
        'system_args':None}
        
args = SimpleNamespace(**args)
E.set_up(args)

convs = E.C.prepare_data(path=args.test_path)

conversations = []

for conv_num, conv in enumerate(convs[:20:2]):
    sns.set_theme(style="white", palette=None)

    if len(conv.utts)>120: continue
        
    focus = []
    entropy = []
    certainty = []

    for utt_num in range(len(conv.utts)):
        print(f'label: {E.C.label_dict[conv.utts[utt_num].label]}')
        words, scores = E.saliency(args, N=50, conv_num=conv_num, k=utt_num)
        words = [i.replace('<s>', '[CLS]') for i in words]
        words = [i.replace('</s>', '[SEP]') for i in words]
        s = [0] + [k for k, word in enumerate(words) if word == '[SEP]']
        turn_scores = [sum(scores[s[i]:s[i+1]]) for i in range(len(s)-1)]
        
        #temp
        plt.rcParams["figure.figsize"] = 10,1

        dist = [abs(i) for i in turn_scores]
        dist = np.array(dist)/sum(dist)
        plt.imshow(dist[np.newaxis,:], cmap="plasma", aspect="auto", vmin=0, vmax=1)
        plt.show()

        focus.append(turn_scores[utt_num])
        entropy.append(saliency_entropy(turn_scores))
        certainty.append(sum(turn_scores))
        
        print(f" conv {conv_num} utt {utt_num}    focus {focus[-1]:.2f}  entropy {entropy[-1]:.2f}")

    conversations.append((focus, entropy, certainty))
    plot_focus(conversations)
    plot_entropy(conversations)
```

```python
plot_focus(conversations)
plot_entropy(conversations)
```
